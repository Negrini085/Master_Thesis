<h1 style="color:red; font-size:70px;"> Mann-Kendall test </h1>

The Mann–Kendall (MK) test is a non-parametric statistical method widely used to detect the presence of a monotonic trend in a time series. It assesses whether data values tend to increase or decrease over time without requiring any assumption about the underlying data distribution. Because of its robustness to outliers and missing values, the MK test is particularly suitable for analyzing environmental and climatic datasets. The test evaluates the null hypothesis of no trend against the alternative hypothesis of an increasing or decreasing trend. One of its key strengths is that it can be applied to data that do not follow a normal distribution. The Mann–Kendall test is commonly used in hydrology to assess trends in rainfall, streamflow, and water quality parameters.

```python
import numpy as np
import pymannkendall as mk
import matplotlib.pyplot as plt
```

<h1 style="color:blue; font-size:50px;"> Classical test </h1>

```python
np.random.seed(42)       

n = 300                    
ampP = 0.3
expP = 0.008
t = np.arange(0, n)

y_exp = np.exp(expP * t)
noise = np.random.normal(0, ampP, size=n)
y = y_exp + noise

test = mk.original_test(y)
print(test)

plt.figure(figsize = (15, 6))
plt.plot(t, y, 'o', label='Data')
plt.plot(t, y_exp, '--', label='Base trend')
plt.title('Upward trend')
plt.ylabel('Value')
plt.xlabel('Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
np.random.seed(42)       

n = 300                    
ampP = 0.3
expP = 0.008
t = np.arange(0, n)

y_x2 = np.power(expP*(300-t), 2)
noise = np.random.normal(0, ampP, size=n)
y = y_x2 + noise

test = mk.original_test(y)
print(test)

plt.figure(figsize = (15, 6))
plt.plot(t, y, 'o', label='Data')
plt.plot(t, y_x2, '--', label='Base trend')
plt.title('Downward trend')
plt.ylabel('Value')
plt.xlabel('Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
np.random.seed(42)       

n = 300                    
ampP = 0.3

yref = np.zeros(n)
noise = np.random.normal(0, ampP, size=n)
y = yref + noise

test = mk.original_test(y)
print(test)

plt.figure(figsize = (15, 6))
plt.plot(t, y, 'o', label='Data')
plt.plot(t, yref, '--', label='Base trend')
plt.title('No trend')
plt.ylabel('Value')
plt.xlabel('Time')
plt.legend()
plt.grid(True)
plt.show()
```
